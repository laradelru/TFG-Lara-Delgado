```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
#300 ITERACIONES

from numpy.core.fromnumeric import resize
func = lambda th: (th[0]-4)**2+4*(th[1]-2)**2

#Representar la funci√≥n

_x =np.linspace(-16,16, 100)
_y =np.linspace(-16,16, 100)
_z=np.zeros((100,100))
for ix, x in enumerate(_x):
  for iy, y in enumerate(_y):
    _z[iy,ix]=func([x,y])

plt.contourf(_x,_y,_z,100)
plt.colorbar()

#Algoritmo NV

Theta = [0,0]
lr=0.01
li= 0.9
plt.plot(Theta[0],Theta[1], "o", c = "white")
grad=np.zeros(2)
g=np.zeros(2)
vmo=[0,0]

for _ in range(300):

  for it, th in enumerate(Theta):
    g[it]= Theta[it] + li*vmo[it]

  grad[0] = 2*g[0]-8
  grad[1]= 8*g[1]-16
  #print(grad)
  
  for it, th in enumerate(Theta):
    vmo[it]= li*vmo[it] - lr* grad[it]
    Theta[it] = Theta[it] + vmo[it]
  plt.plot(Theta[0],Theta[1], ".", c = "red")
  #print(Theta)
plt.plot(Theta[0],Theta[1], "o", c = "green")
print(Theta)
plt.show()

min=[4,2]
Error=np.zeros(2)
for it, th in enumerate(Theta):
  Error[it]=min[it]-Theta[it]
print(Error)
print(np.sqrt(Error[0]**2+Error[1]**2))
```
