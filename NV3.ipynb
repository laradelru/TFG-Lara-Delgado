```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
from numpy.core.fromnumeric import resize
func = lambda th: (th[0]-4)**2+4*(th[1]-2)**2

#Representamos la funci√≥n
_x =np.linspace(-16,16, 100)
_y =np.linspace(-16,16, 100)
_z=np.zeros((100,100))
for ix, x in enumerate(_x):
  for iy, y in enumerate(_y):
    _z[iy,ix]=func([x,y])

plt.contour(_x,_y,_z,100)
plt.colorbar()
#Algoritmo NV
min=[4,2]
Theta = [0,0]
Thetaold=[0,0]
lr=0.1
li= 0.9
plt.plot(Theta[0],Theta[1], "o", c = "white")
grad=np.zeros(2)
g=np.zeros(2)
vmo=[0,0]
contador= 0
diferencia=1

for _ in range(10000):

  if diferencia < 10**-3:
    break
  else:

    for it, th in enumerate(Theta):
      g[it]= Theta[it] + li*vmo[it]

    grad[0] = 2*Theta[0]-8
    grad[1]= 8*Theta[1]-16
    #print(grad)
    for it, th in enumerate(Theta):
      vmo[it]= li*vmo[it] - lr* grad[it]
      Thetaold[it]=Theta[it]
      Theta[it] = Theta[it] + vmo[it]
    plt.plot(Theta[0],Theta[1], ".", c = "red")
    #print(Theta)

    
    diferencia=np.sqrt((min[0]-Theta[0])**2+(min[1]-Theta[1])**2)
    #print(diferencia)
    contador=contador+1

plt.plot(Theta[0],Theta[1], "o", c = "green")
print(Theta)
print(contador)
plt.show()
```
