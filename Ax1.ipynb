```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
#300 ITERACIONES

#funcion
from numpy.core.fromnumeric import resize
func = lambda th: (th[0]-4)**2+4*(th[1]-2)**2

#Representamos la funcion
_x =np.linspace(-16,16, 100)
_y =np.linspace(-16,16, 100)
_z=np.zeros((100,100))
for ix, x in enumerate(_x):
  for iy, y in enumerate(_y):
    _z[iy,ix]=func([x,y])

plt.contourf(_x,_y,_z,100)
plt.colorbar()

#Algorimo AdaMax

Theta = [0,0]
lr=0.01
b1= 0.9
b2=0.999
k=1
E=10**(-8)
plt.plot(Theta[0],Theta[1], "o", c = "white")
grad=np.zeros(2)
m=[0,0]
v=[0,0]
vv=np.zeros(2)
U=np.zeros(2)
M=np.zeros(2)
iteraciones=300

for _ in range(iteraciones):

  grad[0] = 2*Theta[0]-8
  grad[1]= 8*Theta[1]-16
  #print(grad)

  for it, th in enumerate(Theta):
      vv[it]=v[it]
      m[it]= b1*m[it] + (1-b1)* grad[it]
      v[it]= b2*v[it] + (1-b2)* grad[it]**2
  
  b1k= b1**(k)
  b2k= b2**(k)
  k=k+1

  for it, th in enumerate(Theta):

    M[it]= m[it]/(1-b1k)
  
    if b2*vv[0] < abs(grad[0]):
      U[it]=abs(grad[it])
    else:
      U[it]=b2*vv[it]
  
  for it, th in enumerate(Theta):  

    Theta[it] = Theta[it] - lr * M[it]/ U[it]
  #print(Theta)
    plt.plot(Theta[0],Theta[1], ".", c = "red")
print(Theta)
plt.plot(Theta[0],Theta[1], "o", c = "green")

plt.show()

min=[4,2]
Error=np.zeros(2)
for it, th in enumerate(Theta):
  Error[it]=min[it]-Theta[it]
print(Error)
print(np.sqrt(Error[0]**2+Error[1]**2))
```
