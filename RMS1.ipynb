```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
#300 ITERACIONES

from numpy.core.fromnumeric import resize
func = lambda th: (th[0]-4)**2+4*(th[1]-2)**2

#Representamos la funcion
_x =np.linspace(-16,16, 100)
_y =np.linspace(-16,18, 100)
_z=np.zeros((100,100))

for ix, x in enumerate(_x):
  for iy, y in enumerate(_y):
    _z[iy,ix]=func([x,y])

plt.contourf(_x,_y,_z,100)
plt.colorbar()

# Algoritmo RMSProp
Theta = [0,0]
lr=0.01 
li=0.9
E=10**(-8)
plt.plot(Theta[0],Theta[1], "o", c = "white")
grad=np.zeros(2)
Eg = [0,0]

for _ in range(300):

  grad[0] = 2*Theta[0]-8
  grad[1]= 8*Theta[1]-16
  #print(grad)


  for it, th in enumerate(Theta):
   Eg[it]= li*Eg[it]+ (1-li)*grad[it]**2
  #print(Eg)
  for it, th in enumerate(Theta):  

    Theta[it] = Theta[it] - lr * grad[it]/ np.sqrt(Eg[it] + E)
  
    plt.plot(Theta[0],Theta[1], ".", c = "red")
  #print(Theta)

plt.plot(Theta[0],Theta[1], "o", c = "green")
print(Theta)
plt.show()

min=[4,2]
Error=np.zeros(2)
for it, th in enumerate(Theta):
  Error[it]=min[it]-Theta[it]
print(Error)
print(np.sqrt(Error[0]**2+Error[1]**2))

```
